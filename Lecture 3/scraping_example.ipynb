```python
import requests
from bs4 import BeautifulSoup
```

```python
#get html from website
url = 'https://www.vvz.ethz.ch/Vorlesungsverzeichnis/lerneinheit.view?semkez=2025S&ansicht=LEHRVERANSTALTUNGEN&lerneinheitId=186977&lang=en'
response = requests.get(url)
html = response

print(html.text)
```

```python
#prettify html with BeautifulSoup
soup = BeautifulSoup(html.text, 'html.parser')
print(soup.prettify())
```

```python
#retreive date, time and room of lecture 
date = soup.find_all('a', target='Lehrveranstaltung')
time = soup.find_all('a', target='Anfangszeit')
room = soup.find_all('a', target='Rauminfo')

date = ' '.join([elem.text for elem in date])
time = ' '.join([elem.text for elem in time])
room = ' '.join([elem.text for elem in room])

print(date)
print(time)
print(room)
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

def set_chrome():
    #set chrome as webdriver
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    # options.add_argument('disable-gpu')
    service = Service(ChromeDriverManager().install())
    browser = webdriver.Chrome(service=service, options=options)
    return browser

browser = set_chrome()
browser.get(url)
#click on the button with text Learning materials
button = browser.find_element(By.XPATH, "//*[contains(text(), 'Learning materials')]")
button.click()

#get href and text of all elements with target _blank
elements = browser.find_elements(By.XPATH, "//*[@target='_blank']")
for elem in elements:
    print(elem.text + ': ' + elem.get_attribute('href'))
```
